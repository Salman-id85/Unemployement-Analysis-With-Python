```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('Unemployment_Rate_upto_11_2020.csv')  # Replace with your actual dataset path
```

```python

# Data Inspection
print(df.head())
print(df.columns)
```

```python
# Check if 'Date' column exists
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce', format='%d-%m-%Y')  # Adjust format if needed
    df.dropna(subset=['Date'], inplace=True)
else:
    raise ValueError("The dataset does not contain a 'Date' column.")
```

```python
# Strip any leading or trailing whitespace from column names
df.columns = df.columns.str.strip()
```

```python
# Inspect column names
print(df.columns)
```

```python
# Check if 'Date' column exists
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce', format='%Y-%m-%d')  # Adjust format if needed
    df.dropna(subset=['Date'], inplace=True)
else:
    raise ValueError("The dataset does not contain a 'Date' column.")

# Ensure 'UnemploymentRate' column is numeric
if 'latitude' in df.columns:
    df['latitude'] = pd.to_numeric(df['latitude'], errors='coerce')
    df.dropna(subset=['latitude'], inplace=True)
else:
    raise ValueError("The dataset does not contain an 'latitude' column.")
```

```python
# Set 'Date' as the index
df.set_index('Date', inplace=True)
```

```python
# Plot unemployment rate over time
plt.figure(figsize=(14, 7))
plt.plot(df.index, df['latitude'], label='latitude')
plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.legend()
plt.grid(True)
plt.show()
```
